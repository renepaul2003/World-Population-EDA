# World Population 2023- Exploratory Data Analysis

The goal is to gain insights, patterns, and trends in the data through visualizations and statistical analysis.
The project involves Exploratory Data Analysis (EDA) on the world population data from the year 2023 using Python. Through
visualizations and statistical analysis, the aim is to uncover patterns, trends, and insights within the dataset, providing a comprehensive
understanding of global population dynamics during that specific year.

* Two major sources that assess and publish global population are : the United Nations and the U.S. Census Bureau
* The United Nations Population Division of the Department of Economic and Social Affairs every two years calculates, updates, and publishes estimates of total population in its World Population Prospects series
* These population estimates and projections provide the standard and consistent set of population figures that are used throughout the United Nations system

#### Present Dataset contains list of 205 countries including dependent territories

* Dependent territory also known as dependent area, or dependency or an external
territory is a territory that does not possess full political independence or sovereignty as a
sovereign state, yet remains politically outside the controlling state&#39;s integral area
* Rank allocated to each country and dependencies are according to their population size
by United Nations
* Dataset contains nine columns which hold information for each country and dependent
territory

__Columns:__
* country name :   Names of 205 countries including dependent terroitories in the world
* current population :    current population i.e population as of January 2023
* population 2022: population for the previous year i.e. 2022
* area: Area in kilometer square (km^2)
* land area: Land area in km^2
* density: Population density i.e. people per sq. km of land area in /km^2
* growth rate: Annual growth rate where +ve and -ve signs show an increase and
decrease in growth rates respectively
* world percentage: share of each country in total world&#39;s population in terms of
percentage
* rank: rank on the basis of population size, most populated country is assigned highest
rank and so

```python
# changing column world precentage  into float type
df['world percentage']=df['world percentage'].replace("%",' ', regex=True).astype(float)
df['world percentage']
```

```python
# replacing null values with median of the world precentage
world_pre_median=df['world percentage'].median()
df['world percentage'].fillna(world_pre_median,inplace=True)
df.info()
```

```python
# changing column growth rate  into float type
df['growth rate']=df['growth rate'].replace('%',' ',regex=True).astype(float)
df['growth rate']
```

```python
# changing column density into float type
df['density']=df['density'].replace({',': ''},regex=True).astype(float)
df['density']
```

```python
# Function to convert shorthand to numeric
def convert_to_numeric(value):
    if isinstance(value, str):
        value = value.strip().upper()  # Remove leading/trailing spaces and convert to uppercase
        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000  # Convert 'M' to 1 million
        elif 'B' in value:
            return float(value.replace('B', '')) * 1_000_000_000  # Convert 'B' to 1 billion
        elif 'K' in value:
            return float(value.replace('K', '')) * 1_000  # Convert 'K' to 1 thousand
        elif value == '< 1' or value == 'N/A':  # Add conditions for non-numeric cases
            return 0
    return value
df['land area'] = df['land area'].apply(convert_to_numeric)
df['land area'] 
```

```python
# changeing column land area into float type
df['land area'] =df['land area'] .replace({',': ''},regex=True).astype(float)
df['land area'] 
```

```python
# Function to convert shorthand to numeric
def convert_numeric(value):
    if isinstance(value,str):
        value=value.strip().upper() # Remove leading/trailing spaces and convert to uppercase
        if 'M' in value:
            return float(value.replace('M',' ')) * 1_000_000 # Convert 'M' to 1 million
        elif 'B' in value:
            return float(value.replace('B',' ')) * 1_000_000_000 # Convert 'B' to 1 billion
        elif 'K' in value:
            return float(value.replace('K',' ' ))* 1_000 # Convert 'K' to 1 thousand
        elif value == '< 1' or value == 'N/A':  # Add conditions for non-numeric cases
            return 0
    return value
df['area']=df['area'].apply(convert_numeric)
df['area']
```

```python
# looking the correlation of dataset using the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',fmt='.3f')
plt.show()
```

```python
# The countries and dependencies with positive growth rates
postive_growth=df[df['growth rate']>0]
print(postive_growth[['country name','growth rate']])
```

```python
# The countries and dependencies with negative growth rates
neg_growth=df[df['growth rate']<0]
print(neg_growth[['country name','growth rate']])
```

```python
# The countries with 100 million population
countries_100M=df[df['population 2022']>1000000]
print(countries_100M[['country name','population 2022']])
```

```python
# The top five highly densed countries or depedencies in the world
top_5_dense=df.nlargest(5, 'density')
print(top_5_dense[['country name','density']])
```

```python
# The top five least densed countries or depedencies in the world
least_5_dense=df.nsmallest(5, 'density')
print(least_5_dense[['country name','density']])
```

```python
#  ploting the graph of the top 5 densely populated countries
plt.figure(figsize=(12, 6))
# Plot the top 5 densely populated countries
plt.subplot(1, 2, 1)
plt.bar(top_5_dense['country name'], top_5_dense['density'], color='red')
plt.title('Top 5 Densely Populated Countries')
plt.xlabel('Country')
plt.ylabel('Density ')
plt.show()
```

```python
# Plot the least 5 densely populated countries
plt.subplot(1, 2, 2)
plt.bar(least_5_dense['country name'], least_5_dense['density'], color='blue')
plt.title('Least 5  Densely Populated Countries')
plt.xlabel('Country')
plt.ylabel('Density ')
plt.show()
```

```python
#plotting pie chart of the data
plt.figure(figsize=(8, 8))
plt.pie(least_5_dense['country name'], labels=least_5_dense['density'], autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])
```
